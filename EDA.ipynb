# EDA

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv('/content/youtube_subscribers_data.csv')
data
summary_stats = data.describe()
```

```python
plt.figure(figsize=(8, 5))
sns.histplot(data['Subscribers (millions)'], kde=True, color='skyblue', bins=10)
plt.title('Distribution of Subscribers (millions)')
plt.xlabel('Subscribers (millions)')
plt.ylabel('Frequency')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Let's start with summary statistics and distributions. ​​

Observations:
Summary Statistics:

The average number of subscribers is approximately 91 million.
The highest subscriber count is 335 million, while the lowest is 57.5 million.
The distribution is skewed to the right, as evident from the large difference between the mean and maximum.
Subscriber Distribution Plot:

Most channels have between 50 to 100 million subscribers.
There are a few outliers with exceptionally high subscriber counts (e.g., MrBeast, T-Series).

```python
categorical_columns = ['Brand channel', 'Primary language', 'Category']
plt.figure(figsize=(16, 12))
for i, col in enumerate(categorical_columns, 1):
    plt.subplot(2, 2, i)
    sns.countplot(data=data, x=col, palette='viridis', order=data[col].value_counts().index)
    plt.title(f'Count of {col}')
    plt.xticks(rotation=45, ha='right')
    plt.ylabel('Count')
    plt.xlabel(col)

plt.tight_layout()
plt.show()
```

Observations from Categorical Variables:
Brand Channel:

A significant majority of channels are not brand channels.
Non-brand channels dominate the top YouTube subscriber counts.
Primary Language:

The most common language is English, followed by Hindi.
A few other languages, such as Spanish, also appear but less frequently.
Category:

The most represented categories are Entertainment and Music, followed by Education.
These categories seem to attract the largest audiences globally.

```python
plt.figure(figsize=(12, 6))
sns.boxplot(data=data, x='Category', y='Subscribers (millions)', palette='Set2')
plt.title('Subscribers Distribution by Category')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Subscribers (millions)')
plt.xlabel('Category')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```

Observations from Boxplots:
Subscribers by Category:

Music and Entertainment channels have the highest median subscriber counts.
Education channels show significant variation, with some channels achieving very high subscriber counts.
Other categories generally have lower subscriber counts and less variability.
Subscribers by Country:

Channels from India and the United States dominate in terms of subscriber counts.
Indian channels show a higher median compared to most countries, likely due to popular channels like T-Series.
Other countries like Russia and South Korea also feature in the top 10, but with lower variability.

```python
top_countries = data['Country'].value_counts().nlargest(10).index
plt.figure(figsize=(12, 6))
sns.boxplot(data=data[data['Country'].isin(top_countries)], x='Country', y='Subscribers (millions)', palette='coolwarm')
plt.title('Subscribers Distribution by Country (Top 10)')
plt.xticks(rotation=45, ha='right')
plt.ylabel('Subscribers (millions)')
plt.xlabel('Country')
plt.grid(axis='y', linestyle='--', alpha=0.7)
plt.show()
```
