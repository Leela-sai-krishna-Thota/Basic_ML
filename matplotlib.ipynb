1.
Basic Line Plot
o
Plot a simple line graph of ùë¶=ùë•2 for ùë• values ranging from -10 to 10.
o
Label the x-axis as "X Values" and the y-axis as "Squared Values."
o
Add a title to the graph.

```python
import numpy as np
import matplotlib.pyplot as plt


x = np.arange(-10, 11)
y = x**2

plt.plot(x, y)

plt.xlabel("X Values")
plt.ylabel("Squared Values")

plt.title("Plot of y = x^2")

plt.show()
```

3.
Scatter Plot
o
Generate two lists of 20 random integers each (for x and y coordinates).
o
Create a scatter plot with these values, and customize it by adding color, size, and labels to the points.

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.randint(1, 50, 20)
y = np.random.randint(1, 50, 20)
colors = np.random.rand(20)
sizes = 100 * np.random.rand(20)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5, cmap='viridis')
plt.xlabel("X-axis")
plt.ylabel("Y-axis")
```

4. Histogram o Simulate 100 random values from a normal distribution using np.random.randn() and plot a histogram. o Customize the number of bins and add a grid to the plot.

```python
import numpy as np
import matplotlib.pyplot as plt

data = np.random.randn(100)

plt.hist(data, bins=20)

plt.grid(True)

plt.xlabel("Values")
plt.ylabel("Frequency")
plt.title("Histogram of Normal Distribution")

plt.show()
```

5.
Pie Chart
o
Visualize the percentage distribution of five categories ['A', 'B', 'C', 'D', 'E'] with values [20, 30, 25, 15, 10].
o
Display the percentage values on the chart and use a color palette for differentiation

```python
import matplotlib.pyplot as plt

categories = ['A', 'B', 'C', 'D', 'E']
values = [20, 30, 25, 15, 10]

colors = ['#ff9999','#66b3ff','#99ff99','#ffcc99', '#c2c2f0']

plt.pie(values, labels=categories, colors=colors, autopct='%1.1f%%', startangle=90)

plt.title("Percentage Distribution of Categories")

plt.axis('equal')

plt.show()
```

7.
Heatmap (Seaborn)
o
Create a 5x5 matrix with random integers between 1 and 10 using np.random.randint() and visualize it as a heatmap using Seaborn.
o
Add a color bar and annotate the heatmap with the values.

```python
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt

data = np.random.randint(1, 11, size=(5, 5))

sns.heatmap(data, annot=True, cmap="viridis", fmt="d", linewidths=.5)

plt.title("Heatmap of Random Integers")

plt.show()
```

9.
Pair Plot (Seaborn)
o
Manually create a small dataset with columns such as 'Age', 'Height', 'Weight' for 10 individuals and visualize the relationships using a pair plot.

```python
import seaborn as sns
import pandas as pd
import matplotlib.pyplot as plt

# Create a sample dataset
data = {'Age': [25, 30, 22, 28, 35, 20, 27, 32, 29, 24],
        'Height': [175, 180, 165, 172, 185, 160, 170, 182, 178, 168],
        'Weight': [70, 75, 60, 68, 80, 55, 65, 78, 72, 62]}
df = pd.DataFrame(data)

# Create the pair plot
sns.pairplot(df)

# Add a title
plt.suptitle("Pair Plot of Age, Height, and Weight", y=1.02)  # Adjust y for title position

# Display the plot
plt.show()
```

10.
Save and Style Plots
‚Ä¢
Plot any graph of your choice and save it in both .png and .pdf formats.
‚Ä¢
Apply at least one built-in style (e.g., 'ggplot' or 'seaborn-darkgrid') and explain how the style impacts the visualization.

```python
import matplotlib.pyplot as plt
import numpy as np

# Data
categories = ['A', 'B', 'C', 'D', 'E']
values = [25, 30, 20, 15, 10]

plt.style.use('ggplot')

plt.bar(categories, values)

plt.xlabel("Categories")
plt.ylabel("Values")
plt.title("Styled Bar Chart")

plt.savefig("bar_chart.png")
plt.savefig("bar_chart.pdf")

plt.show()
```
